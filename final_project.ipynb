# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

# Plotting the histogram for the 'ConvertedComp' column
QUERY = """
SELECT ConvertedComp
FROM master
"""

converted_comp_data = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.hist(converted_comp_data['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


```python
# Query to get the count of respondents who currently work with 'SQL'
query = """
SELECT COUNT(*) as count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith = 'SQL'
"""

# Read the data into a DataFrame
df_sql_count = pd.read_sql_query(query, conn)
sql_count = df_sql_count['count'][0]

print(f"The number of respondents who currently work with SQL is: {sql_count}")
```

```python
# Query to find the count of respondents who work with 'MySQL' only
query = """
SELECT COUNT(*) as count
FROM (
    SELECT Respondent, GROUP_CONCAT(LanguageWorkedWith) as Languages
    FROM LanguageWorkedWith
    GROUP BY Respondent
    HAVING Languages = 'MySQL'
)
"""

# Read the data into a DataFrame
df_mysql_only_count = pd.read_sql_query(query, conn)
mysql_only_count = df_mysql_only_count['count'][0]

print(f"The number of respondents who work on MySQL only is: {mysql_only_count}")
```

Plot a box plot of `Age.`


```python
# your code goes here
# Query to retrieve the 'Age' column from the 'master' table
QUERY = """
SELECT Age
FROM master
"""

# Retrieving data from the database
age_data = pd.read_sql_query(QUERY, conn)

# Plotting the box plot for the 'Age' column
plt.figure(figsize=(8, 6))
plt.boxplot(age_data['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# Query to retrieve the 'Age' and 'WorkWeekHrs' columns from the 'master' table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Retrieving data from the database
age_workweek_data = pd.read_sql_query(QUERY, conn)

# Plotting the scatter plot for 'Age' and 'WorkWeekHrs'
plt.figure(figsize=(8, 6))
plt.scatter(age_workweek_data['Age'], age_workweek_data['WorkWeekHrs'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Query to retrieve the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' columns from the 'master' table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Retrieving data from the database
work_code_age_data = pd.read_sql_query(QUERY, conn)

# Plotting the bubble plot for 'WorkWeekHrs' and 'CodeRevHrs' with 'Age' as bubble size
plt.figure(figsize=(10, 8))
plt.scatter(work_code_age_data['WorkWeekHrs'], work_code_age_data['CodeRevHrs'], s=work_code_age_data['Age']*5, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
# Connect to the SQLite database
# Query to get the counts of databases respondents wish to learn next year
query = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

# Read the data into a DataFrame
df_databases = pd.read_sql_query(query, conn)

# Plot pie chart
plt.figure(figsize=(10, 6))
plt.pie(df_databases['count'], labels=df_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Query to get the counts of languages respondents wish to learn next year
query = """
SELECT LanguageDesireNextYear, COUNT(*) as count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
"""

# Read the data into a DataFrame
df_languages = pd.read_sql_query(query, conn)

# Find the rank of Python
python_rank = df_languages[df_languages['LanguageDesireNextYear'] == 'Python'].index[0] + 1

print(f"The rank of Python is: {python_rank}")
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the 'Age' and 'ConvertedComp' columns from the 'master' table
QUERY = """
SELECT Age, ConvertedComp
FROM master
"""

# Retrieving data from the database
age_comp_data = pd.read_sql_query(QUERY, conn)

# Filter data for ages between 45 and 60
age_comp_data_filtered = age_comp_data[(age_comp_data['Age'] >= 45) & (age_comp_data['Age'] <= 60)]

# Group data by 'Age' and calculate the median 'ConvertedComp' for each age
median_comp_by_age = age_comp_data_filtered.groupby('Age')['ConvertedComp'].median()

# Plotting the median ConvertedComp for ages from 45 to 60
plt.figure(figsize=(10, 6))
median_comp_by_age.plot(marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the 'MainBranch' column from the 'master' table
QUERY = """
SELECT MainBranch
FROM master
"""

# Retrieving data from the database
main_branch_data = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each category in the 'MainBranch' column
main_branch_counts = main_branch_data['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x')
plt.show()
```

```python
# Query to find the country with the most responders
query = """
SELECT Country, COUNT(*) as count
FROM master
GROUP BY Country
ORDER BY count DESC
LIMIT 1
"""

# Read the data into a DataFrame
df_majority_country = pd.read_sql_query(query, conn)
majority_country = df_majority_country['Country'][0]
majority_count = df_majority_country['count'][0]

print(f"The majority of survey responders are from {majority_country} with {majority_count} responders.")
```

```python
# Query to find the most common developer type
query = """
SELECT DevType, COUNT(*) as count
FROM DevType
GROUP BY DevType
ORDER BY count DESC
LIMIT 1
"""

# Read the data into a DataFrame
df_majority_devtype = pd.read_sql_query(query, conn)
majority_devtype = df_majority_devtype['DevType'][0]
majority_count = df_majority_devtype['count'][0]

print(f"The majority of survey responders are {majority_devtype} with {majority_count} responders.")
```
